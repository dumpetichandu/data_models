```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from  sklearn import linear_model
```

```python
df = pd.read_csv('linear regression.csv')
df.head(5)
```

```python
%matplotlib inline
plt.scatter(df.year,df.income)
```

```python
r = linear_model.LinearRegression()
```

```python
r.fit(df[['year']],df.income)
```

```python
int(r.predict([[2023]]))
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('income',fontsize=20)
plt.scatter(df.year,df.income,color='red')
plt.plot(df.year,r.predict(df[['year']]),color='blue')
```

```python
r.coef_
```

```python
import pickle as p
with open('Prediction _the_income_model','wb') as f:
    p.dump(r,f)
```

```python
with open('Prediction _the_income_model','rb') as f:
    model=p.load(f)
```

```python
model.predict([[2023]])
```
